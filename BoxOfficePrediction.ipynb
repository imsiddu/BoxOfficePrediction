```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv(r"C:\Users\siddu\OneDrive\Desktop\Movie-Box-Office-Prediction-main\3.1 cost_revenue_clean.csv.csv")
```

```python
data.describe()
```

```python
X = DataFrame(data,columns=['production_budget_usd'])
y = DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,alpha=0.3)
plt.title('Production Budget x Gross Collections')
plt.xlabel('Production Budget in $')
plt.ylabel('Gross Collections in $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

Plot :

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,alpha=0.3)
plt.plot(X, reg.predict(X), color='red', linewidth=5)
plt.title('Production Budget x Gross Collections')
plt.xlabel('Production Budget in $')
plt.ylabel('Gross Collections in $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
reg.score(X,y)
```
